```python
import pandas as pd
import csv
conflict_data = pd.read_csv('Resources/ged191.csv')
import numpy as np
import matplotlib.pyplot as plt
# just show the ~4 early 90's significant trends and 
#the ~4 late 2018ish significant trends and don't show the other trends.
```

```python
print (conflict_data)
```

```python
conflict_data.head()
```

```python
 df = conflict_data
```

```python
df.keys()
```

```python
df['country'].unique()
```

```python
df_countries = df [ (df['country'] == 'Afghanistan') | (df['country'] == 'Ethiopia') | (df['country'] == 'Iraq') |  (df['country'] == 'Congo') |  (df['country'] == 'Angola') |  (df['country'] == 'Nigeria') |  (df['country'] == 'Sri Lanka')]
df_countries

```

```python
new2_df = df_countries[['country', 'year','best']].groupby(['country','year']).sum()
new2_df

```

```python
df_now = new2_df.reset_index()
df_now.head()
```

```python
afg_year = df_now[df_now['country']=='Afghanistan']['year']
afg_best = df_now[df_now['country']=='Afghanistan']['best']
plt.plot(afg_year, afg_best)
plt.title('Deaths over Time Afghanistan')
plt.xlabel('Year')
plt.ylabel('Deaths')
```

```python
eth_year = df_now[df_now['country']=='Ethiopia']['year']
eth_best = df_now[df_now['country']=='Ethiopia']['best']
plt.plot(eth_year, eth_best)
plt.title('Deaths over Time Ethiopia')
plt.xlabel('Year')
plt.ylabel('Deaths')
```

```python
ang_year = df_now[df_now['country']=='Angola']['year']
ang_best = df_now[df_now['country']=='Angola']['best']
plt.plot(ang_year, ang_best)
plt.title('Deaths over Time Angola')
plt.xlabel('Year')
plt.ylabel('Deaths')
```

```python
sri_year = df_now[df_now['country']=='Sri Lanka']['year']
sri_best = df_now[df_now['country']=='Sri Lanka']['best']
plt.plot(sri_year, sri_best)
plt.title('Deaths over Time Sri Lanka')
plt.xlabel('Year')
plt.ylabel('Deaths')
```

```python
nig_year = df_now[df_now['country']=='Nigeria']['year']
nig_best = df_now[df_now['country']=='Nigeria']['best']
plt.plot(nig_year, nig_best)
plt.title('Deaths over Time Nigeria')
plt.xlabel('Year')
plt.ylabel('Deaths')
```

```python
irq_year = df_now[df_now['country']=='Iraq']['year']
irq_best = df_now[df_now['country']=='Iraq']['best']
plt.plot(irq_year, irq_best)
plt.title('Deaths over Time Iraq')
plt.xlabel('Year')
plt.ylabel('Deaths')
```

```python
ax=plt.figure(figsize=(20, 10))
plt.grid()
plt.plot(afg_year, afg_best, linewidth=7, label='Afghanistan')
plt.plot(eth_year, eth_best, linewidth=3.5, label='Ethiopia')
plt.plot(ang_year, ang_best, linewidth=3.5, label='Angola')
plt.plot(sri_year, sri_best, linewidth=3.5, label='Sri Lanka')
plt.plot(nig_year, nig_best, linewidth=3.5, label='Nigeria')
plt.plot(irq_year, irq_best,linewidth=3.5, label='Iraq')
ax.legend(loc='upper right', borderpad=2, fontsize=16, labelspacing=1.2)

plt.title('Fatalities by Countries with Highest Death Rates (excluding Rwanda)', fontsize=18)
plt.xlabel('Year', fontsize=18)
plt.ylabel('Deaths', fontsize=18)
plt.savefig('Fatalities by Countries with Highest Death Rates.png')
plt.show()

```
